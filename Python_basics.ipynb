```python
# Day 1 – Python Basics

# Σε αυτό το notebook μαθαίνουμε:
# Μεταβλητές
# Τύπους δεδομένων
# Πράξεις με αριθμούς
# Strings
# Input/Output
# Mini project: Υπολογισμός μέσου όρου βαθμών

print("Hello, AI World!")

#Μεταβλητές
name = "thodoris"
age = 25
height = 1.70
is_student = True

print(name, age, height, is_student)
print(type(name))
print(type(age))

#Πράξεις με Αριθμούς
a = 10
b = 3

print(a + b)   # πρόσθεση
print(a - b)   # αφαίρεση
print(a * b)   # πολλαπλασιασμός
print(a / b)   # διαίρεση
print(a // b)  # ακέραιη διαίρεση
print(a % b)   # υπόλοιπο
print(a ** b)  # δύναμη

#Strings
text = "Hello AI"
print(text.upper())
print(text.lower())
print(text.replace("AI", "World"))
print(text[0])
print(text[-1])

#Input Χρήστη
name = input("Πώς σε λένε; ")
age = int(input("Πόσο χρονών είσαι; "))
print(f"Γεια σου {name}, σε 5 χρόνια θα είσαι {age + 5}!")

#Υπολογισμός Μέσου Όρου βαθμών
grade1 = float(input("Βαθμός 1: "))
grade2 = float(input("Βαθμός 2: "))
grade3 = float(input("Βαθμός 3: "))

average = (grade1 + grade2 + grade3) / 3

print("Ο μέσος όρος είναι:", average)

if average >= 5:
    print("Πέρασες το μάθημα!")
else:
    print("Δυστυχώς κόπηκες.")

print("=== Student Info System ===")

name = input("Όνομα: ")
age = int(input("Ηλικία: "))
grades = []

for i in range(3):
    grade = float(input(f"Βαθμός {i+1}: "))
    grades.append(grade)

average = sum(grades) / len(grades)

print("\n--- Αναφορά ---")
print(f"Μαθητής: {name}")
print(f"Ηλικία: {age}")
print(f"Βαθμοί: {grades}")
print(f"Μέσος Όρος: {average:.2f}")

if average >= 5:
    print("Κατάσταση: ✅ Επιτυχία")
else:
    print("Κατάσταση: ❌ Αποτυχία")

```
